# Uso de matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[255, 0, 90], [0, 90, 0]])
x
```

```python
plt.figure(figsize=(5, 5))
plt.subplot(2,2,1)
plt.imshow(x, vmin=0, vmax=255, cmap='gray')
plt.title('Escala de grises')
plt.subplot(2,2,2)
plt.imshow(x, vmin=0, vmax=255, cmap='viridis')
plt.title('Viridis')
plt.subplot(2,2,3)
plt.imshow(x, vmin=0, vmax=255, cmap='magma')
plt.title('Magma')
```
