# Uso de matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[255, 0, 90], [0, 90, 0]])
x
```

```python
plt.figure(figsize=(5, 5))
plt.subplot(2,2,1)
plt.imshow(x, vmin=0, vmax=255, cmap='gray')
plt.title('Escala de grises')
plt.subplot(2,2,2)
plt.imshow(x, vmin=0, vmax=255, cmap='viridis')
plt.title('Viridis')
plt.subplot(2,2,3)
plt.imshow(x, vmin=0, vmax=255, cmap='magma')
plt.title('Magma')
```

```python
Ir = np.array([[255, 0, 0],[0, 255, 128]])
Ig = np.array([[0, 255, 0],[0, 255, 20]])
Ib = np.array([[0, 0, 255],[0, 255, 100]])
Irgb = np.stack((Ir, Ig, Ib), axis=2)
```

```python
Irgb.shape
```

```python
Irgb[:,:,2]
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(Irgb, vmin=0, vmax=255)
```

```python
# Cargar librer√≠a OpenCV
import cv2
```

```python
# Cargar imagen con opencv
x = cv2.imread("data/arbol.png")
```

```python
x.shape
```

```python
# Convertir BGR a RGB
x_rgb = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
```

```python
x_rgb.shape
```

```python
x_rgb[0,0,:]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 5))
plt.subplot(1,2,1)
plt.imshow(x)
plt.axis('off')
plt.title('BGR')
plt.subplot(1,2,2)
plt.imshow(x_rgb)
plt.axis('off')
plt.title('RGB')
```

```python
x_crop_vert = x_rgb[:,0:384,:]
x_crop_vert.shape
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(x_crop_vert)
plt.axis('off')
```

```python
x_crop_vert_right = x_rgb[:,384:,:]
plt.figure(figsize=(5, 5))
plt.imshow(x_crop_vert_right)
plt.axis('off')
```

```python
x_concat = np.concatenate((x_crop_vert, x_crop_vert_right), axis=1)
plt.figure(figsize=(5, 5))
plt.imshow(x_concat)
plt.axis('off')
```

```python
x_blue = x_rgb.copy()
x_blue[:,:,0] = 0
x_blue[:,:,1] = 0
plt.figure(figsize=(5, 5))
plt.imshow(x_blue)
plt.title('Canal Azul')
plt.axis('off')
```
